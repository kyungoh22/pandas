```python
# which Genre is favored the most by the people?

# If you judge by rating: 
movies_df.groupby('genre').mean()[['rating']].sort_values('rating', ascending = False)
```

```python
# If you judge by ticket sales: 
movies_df.groupby('genre').mean()[['gross']].sort_values('gross', ascending = False)
```
